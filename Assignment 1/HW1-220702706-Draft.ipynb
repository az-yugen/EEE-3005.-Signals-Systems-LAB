```python
# defining some elementary signals for repeated use

# unit step
def step(t,pos):
    return np.heaviside(t+pos,1)

# unit impulse
# def impulse(t, pos):
#     sig = []
#     for i in t:
#         if i == pos:
#             sig.append(1)
#         else:
#             sig.append(0)
#     return sig

# def impulse(t, pos):
#    sig = np.zeros(t)
#    sig[pos] = 1
#    return sig

def impulse(t, pos):
    return np.where(t == -pos, 1, 0)


# unit ramp
def ramp(t):
    sig = []
    for i in range(len(t)):
        value = (t[i] if t[i]>=0 else 0)
        sig.append(value)
    return sig

# rectangular pulse
def rect_pulse(t, tau):
    return np.heaviside(t,1) - np.heaviside(t-tau,1)

```

```python

# defined time interval for 1st question
t = np.linspace(-1,5,1000)



# INITIALIZING SIGNALS a to h.

u_t = step(t,0)
r_t = ramp(t)
x_1c = 3*np.exp(-2*t)*u_t

def x_1d(t):
    signal = []
    for i in range(len(t)):
        if t[i] >= 0 and t[i]<=2:
            value = np.exp(-t[i])
        else:
            value = 0
        #value = (np.exp(-t[i]) if t[i] >= 0 && t[i]<=2 else 0)
        signal.append(value)
    return signal

x_1e = np.sin(2*t) + 2*np.cos(3*t-0.2)
x_1f = rect_pulse(t, 2)
x_1g = np.exp(2*t)*np.sin(3*t)*u_t
x_1h = np.exp(-2*t)*np.sin(3*t)*u_t



# PREPARING PLOT FIGURE

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(12, 6))
fig.tight_layout(pad=4)
plt.rc('axes.spines', **{'bottom':True, 'left':True, 'right':False, 'top':False})
plt.rcParams.update({"text.usetex": True})



# PLOTTING

# Part a
ax1 = plt.subplot(241)
ax1.plot(t,u_t)
plt.title('$(a): u(t)$')

# Part b
ax2 = plt.subplot(242)
ax2.plot(t,r_t)
plt.title('$(b): r(t)$')

# Part c
ax3 = plt.subplot(243)
ax3.plot(t,x_1c)
plt.title('$(c): 3e^{-2t}u(t)$')

# Part d
ax4 = plt.subplot(244)
ax4.plot(t, x_1d(t))
plt.title('$(d): e^{-t}, 0 \leq t \leq 2$')

# Part e
ax5 = plt.subplot(245)
ax5.plot(t,x_1e)
plt.title('$(e): sin(2t)+2cos(3t-0.2)$')

# Part f
ax6 = plt.subplot(246)
ax6.plot(t, x_1f)
plt.title('$(f):p_2(t)$')

# Part g
ax7 = plt.subplot(247)
ax7.plot(t,x_1g)
plt.ylim(-30,30) # the function has very abrupt fluctuations so i scaled y-axis down, so it could be better viewed
plt.title('$(g): e^{2t}sin(3t)u(t)$')

# Part h
ax8 = plt.subplot(248)
ax8.plot(t,x_1h)
plt.title('$(h): e^{-2t}sin(3t)u(t)$')
```
